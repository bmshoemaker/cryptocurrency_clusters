```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
# Read crypto_data.csv into Pandas.
file = 'Resources/crypto_data.csv'
```

```python
df = pd.read_csv(file)
df.head()
```

```python
# Discard all cryptocurrencies that are not being traded.
crypto_df = df[df['IsTrading'] == True]
crypto_df.head()
```

```python
# drop the IsTrading column from the dataframe
crypto_filt = crypto_df.drop(columns='IsTrading')
crypto_filt.head()
```

```python
print(len(crypto_filt))
```

```python
# Remove all rows that have at least one null value.
crypto_filt2=crypto_filt.dropna()
crypto_filt2.head()
```

```python
print(len(crypto_filt2))
```

```python
# Filter for cryptocurrencies that have been mined
mined_df = crypto_filt2[crypto_filt2['TotalCoinsMined'] > 0]
mined_df.head()
```

```python
print(len(mined_df))
```

```python
# delete the CoinName from the original dataframe
clean_df = mined_df.drop(columns='CoinName')
clean_df.head()
```

```python
# Convert the remaining features with text values, Algorithm and ProofType, into numerical data using dummy variables
dumm_df = pd.get_dummies(clean_df[['Algorithm', 'ProofType']])
dumm_df = clean_df.merge(dumm_df, left_index=True, right_index=True)
dumm_df
```

```python
dumm_df = dumm_df.drop(columns=['Unnamed: 0', 'Algorithm', 'ProofType'])
dumm_df
```

```python
# Standardize your dataset
scaler = StandardScaler().fit(dumm_df)
scaled_df = scaler.transform(dumm_df)
scaled_df
```

```python
# Perform dimensionality reduction with PCA. Preserve 90% of the explained variance in dimensionality reduction. 
# Initialize PCA model
pca = PCA(n_components=.9)
# Get principal components for the data.
crypto_pca = pca.fit_transform(scaled_df)
# Transform PCA data to a DataFrame
crypto_pca=pd.DataFrame(crypto_pca)
crypto_pca.head()
```

```python
# Run t-SNE on the principal components
 # Initialize t-SNE model
tsne = TSNE(learning_rate=100)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(crypto_pca)
```

```python
# The dataset has 2 columns
tsne_features.shape
```

```python
# Create a scatter plot of the t-SNE output. Observe whether there are distinct clusters or not
# The first column of transformed features
crypto_pca['x'] = tsne_features[:,0]

# The second column of transformed features
crypto_pca['y'] = tsne_features[:,1]
```

```python
 # Visualize the clusters
plt.scatter(crypto_pca['x'], crypto_pca['y'])
plt.show()
```

```python
### clusters can be identified in the scatter plot
```

```python
# Create an elbow plot to identify the best number of clusters
# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
# The elbow of the plot is ~ at 5 number of clusters and 75,000 inertia
```

```python
### Recommendation
# It looks as though the data can potentially be clustered together. The optimal number of clusters would be 5. 
```
